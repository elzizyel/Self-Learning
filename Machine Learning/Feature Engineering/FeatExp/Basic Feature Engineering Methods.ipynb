##  无量纲化

###  标准化（Standardization）
    使得所有数据点缩放到符合mean=0 & σ = 1

    若原始数据点为Normal distribution, 那么scale完的就是N(0, 1)
\begin{equation*}
x' = \frac{x - \bar{X}}{\sigma_x}
\end{equation*}

```python
from sklearn.preprocessing import StandardScaler

StandardScaler().fit_transform(x)
```

### 区间缩放法 (Min-Max Scaling)
    对原始数据进行线性变化, 使其落入[0,1]区间内
\begin{equation*}
x' = \frac{x - Min_x}{Max_x - Min_x}
\end{equation*}

```python
from sklearn.preprocessing import MinMaxScaler

MinMaxScaler().fit_transform(x)
```

### L2正则化 (L2 Normalization / Euclidean Normalization)
    依照特征矩阵进行行处理, 使得样本向量在点乘/核函数计算时, 拥有统一的标准，即转化为”单位向量“
    
    1.1 / 1.2 是依照特征矩阵进行列处理
\begin{equation*}
x' = \frac{x}{||x||_2}
\end{equation*}

\begin{equation*}
||x||_2 = \sqrt{x_1^2 + x_2^2 + ... + x_m^2}
\end{equation*}

<img src="l2_normalization.png" width="40%">

```python
from sklearn.preprocessing import Normalizer

Normalizer().fit_transform(x)
```

### 定量特征二值化 (Binarization)
    设定一个阈值, 大于该阈值x为1, 小于该阈值x为0

\begin{equation*}
x' = \frac{1, x > threshold}{0, x ≤ threshold}
\end{equation*}


```python
from sklearn.preprocessing import Binarizer

Binarizer(threshold=3).fit_transform(x)
```

## 特征选择

- Filter: 按照发散性或者相关性对各个特征进行评分, 设定阈值或者带选择阈值的个数, 选择特征
- Wrapper: 根据目标函数，每次选择若干特征/排除若干特征
- Embedded: 先用模型训练，得到各个特征的权重，根据权重选择特征

### Filter

#### 方差选择法 (Variance Thresholding)
    计算各个特征的方差, 根据阈值, 选择方差大于阈值的特征

```python
from sklearn.feature_selection import VarianceThreshold

VarianceThreshold(threshold=3).fit_transform(x)
```

#### 相关系数法
    计算各个特征对target的相关系数(r)以及相关系数的p-value
    然后进行选择

```python
from sklearn.feature_selection import SelectKBest
from scipy.stats import pearsonr

# SelectKbest内为评价指标, k=2 为选出最好的2个
# fit_transform内为x, y
SelectKBest(lambda X, Y: np.array(list(map(lambda x: pearsonr(x, Y)[0], X.T))).T, k=2).fit_transform(x, y)
```

#### 卡方检验
    通过自变量和因变量之间的相关性

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#选择K个最好的特征，返回选择特征后的数据
SelectKBest(chi2, k=2).fit_transform(x, y)
```

#### 互信息法
    评价定性x对定性y的相关性

```python
from sklearn.feature_selection import SelectKBest
from minepy import MINE
 
#由于MINE的设计不是函数式的，定义mic方法将其为函数式的，返回一个二元组，二元组的第2项设置成固定的P值0.5
def mic(x, y):
 m = MINE()
 m.compute_score(x, y)
 return (m.mic(), 0.5)

#选择K个最好的特征，返回特征选择后的数据
SelectKBest(lambda X, Y: np.array(list(map(lambda x: mic(x, Y)[0], X.T))).T, k=2).fit_transform(x, y)
```

### Wrapper

#### 递归特征消除法
    使用一个模型进行多伦训练, 每轮训练后, 消除权重过小的特征

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#递归特征消除法，返回特征选择后的数据
#参数estimator为基模型
#参数n_features_to_select为选择的特征个数
RFE(estimator=LogisticRegression(), n_features_to_select=2).fit_transform(x, y)
```

### Embedded

#### 基于树模型的特征选择法
    例子中使用GBDT来做特征选择

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import GradientBoostingClassifier

#GBDT作为基模型的特征选择
SelectFromModel(GradientBoostingClassifier()).fit_transform(x, y)
```

## 降维
    当特征选择完成后, 其实就可以训练模型了, 但是由于feature矩阵过大, 训练时间会很长。
    所以可以对feature先进行降维, 再进行训练。

### PCA

```python
from sklearn.decomposition import PCA

#主成分分析法，返回降维后的数据
#参数n_components为主成分数目
PCA(n_components=2).fit_transform(x)
```

### LDA

```python
from sklearn.decomposition import LatentDirichletAllocation

#线性判别分析法，返回降维后的数据
#参数n_components为降维后的维数
LatentDirichletAllocation(n_components=2).fit_transform(x, y)
```
